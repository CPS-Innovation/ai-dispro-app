```python
import pandas as pd
from IPython.display import display, clear_output
import ipywidgets as widgets

def prompt_entry_ui(df=None):
    """
    Creates an interactive ipywidgets UI with a single data entry field 'prompt'.
    If an existing DataFrame is provided, it will append to it.
    If no DataFrame is provided, a new one will be created.
    
    Returns
    -------
    get_df : function
        Call get_df() to retrieve the updated DataFrame at any time.
    """

    # --- Initialise df ---
    if df is None:
        df = pd.DataFrame(columns=["prompt"])

    # --- Widgets ---
    prompt_box = widgets.Textarea(
        value="",
        placeholder="Type your prompt here...",
        description="Prompt:",
        layout=widgets.Layout(width="100%", height="150px")
    )

    save_btn = widgets.Button(
        description="Save",
        button_style="success",
        icon="save"
    )

    status = widgets.HTML(value="")
    out_table = widgets.Output()

    # --- Save handler ---
    def on_save_clicked(_):
        text = prompt_box.value.strip()

        if not text:
            status.value = "<span style='color:#d43f3a;'>Please enter a prompt before saving.</span>"
            return

        nonlocal df
        df = pd.concat([df, pd.DataFrame([{"prompt": text}])], ignore_index=True)

        prompt_box.value = ""
        status.value = "<span style='color:#3c763d;'>Saved!</span>"
        refresh_table()

    save_btn.on_click(on_save_clicked)

    # --- Refresh table ---
    def refresh_table():
        with out_table:
            clear_output(wait=True)
            display(df.tail(10) if not df.empty else pd.DataFrame({"prompt": []}))

    # --- UI Layout ---
    ui = widgets.VBox([
        widgets.HTML("<h3>Enter Prompt</h3>"),
        prompt_box,
        save_btn,
        status,
        widgets.HTML("<hr>"),
        widgets.HTML("<b>Recent Entries:</b>"),
        out_table
    ])

    display(ui)
    refresh_table()

    # --- Return accessor ---
    return lambda: df
```

```python
get_df = prompt_entry_ui()
```

```python
output_file = "../data/prompts.xlsx"

df = get_df()

df.to_excel(
    output_file,
    index=False,
    engine="openpyxl"
)
```
